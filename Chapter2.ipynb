```python
#The volume of a sphere with radius r is 4/3πr**3 . What is the volume of a sphere with radius 5?
import math
radius = 5
volume = (4/3)*math.pi*radius**3
print(volume)
```

```python
#If I leave my house at 6:52 am and run 1 mile at an easy pace (8:15 per mile), 
#then 3 miles at tempo (7:12 per mile) and 1 mile at easy pace again, 
#what time do I get home for breakfast?
```

```python
import datetime

total_minutes_run = 1*8.25 + 3*7.20 + 1*8.25

print("Total morning Run: ",total_minutes_run)

time = datetime.datetime(1,1,1,6,52,0)

new_time = time + datetime.timedelta(minutes=total_minutes_run)

print("I get home for breakfast: ",new_time)
```
